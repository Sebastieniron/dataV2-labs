## Challenge 1
I work at a coding bootcamp, and I have developed a theory that the younger my students are, the more often they are late to class. In order to test my hypothesis, I have collected some data in the following table:

| StudentID | Age | Tardies |
|--------|-----|------------|
| 1      | 17  | 10         |
| 2      | 51  | 1          |
| 3      | 27  | 5          |
| 4      | 21  | 9         |
| 5      | 36  |  4         |
| 6      | 48  |  2         |
| 7      | 19  |  9         |
| 8      | 26  | 6          |
| 9      | 54  |  0         |
| 10     | 30  |  3         |

Use this command to create a dataframe with the data provided in the table. 
~~~~
student_data = pd.DataFrame({'X': [x_values], 'Y': [y_values]})
~~~~

```python
# Your code here. 
import pandas as pd
student_data = pd.DataFrame({'X': [17,51,27,21,36,48,19,26,54,30], 'Y': [10,1,5,9,4,2,9,6,0,3]})
student_data
```

Draw a dispersion diagram (scatter plot) for the data.

```python
# Your code here.
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(student_data['X'],student_data['Y'])
```

Do you see a trend? Can you make any hypotheses about the relationship between age and number of tardies?

```python
# Your response here. 

#The older people are the more on time they are.
```

Calculate the covariance and correlation of the variables in your plot. What is the difference between these two measures? Compare their values. What do they tell you in this case? Add your responses as comments after your code.

```python
# Your response here.
corr = student_data.corr()
cov = student_data.cov()
print(corr)
print(cov)
#We can clearly see that age & tardies are negatively correlated.
#The difference between the two is that thet correlation represents the covariance but on a scale of -1 to 1.
#Covariance shows the direction of the linear relationship while correlation both shows strength & direction of the 
#linear relationship.
```

Build a regression model for this data. What will be your outcome variable? What type of regression are you using? Add your responses as comments after your code.

```python
# Your response here.
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(student_data['X'].values.reshape(-1,1), student_data['Y'].values.reshape(-1,1))

intercept = model.intercept_
slope = model.coef_
print(intercept)
print(slope)

intercept = model.intercept_[0]
slope = model.coef_[0][0]
print(intercept)
print(slope)

#Here we calculated the slope or rate of change of the trendline which indicates our negative relationship since the slope is
#negative. We also calculated the intercept which shows when the trendline crosses the y-axis so when x would be zero.
#Meaning that if a student were to be 0 of age he would be 12 times late, in context this is impossible due to age.
```

Plot your regression model on your scatter plot.

```python
# Your code here.
import seaborn as sns
sns.regplot(x="X", y="Y", data=student_data)
```

Interpret the results of your model. What can conclusions can you draw from your model and how confident in these conclusions are you? Can we say that age is a good predictor of tardiness? Add your responses as comments after your code.

```python
# Your response here. 
predictions = [intercept + slope*x for x in student_data['X']]
predictions
```

```python
residuals = [student_data['Y'][i] - predictions[i] for i in range(len(student_data['Y']))]
residuals = [realvalue - predictedvalue for realvalue, predictedvalue in zip(student_data['Y'],predictions)]
residuals
```

```python
import numpy as np
MSE = np.mean([residual**2 for residual in residuals])
print('MSE = ' + str(MSE))
RMSE = np.sqrt((np.mean([residual**2 for residual in residuals])))
RMSE = np.sqrt(MSE)
print('RMSE = ' + str(RMSE))
Per_error = (RMSE / student_data['Y'].mean()) * 100
print('% error = ' + str(Per_error))

#
```
