```python
import pandas as pd
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = 'C:/Users/svvan/Desktop/dataV2-labs/module-2/ProjectWeek6/DF_Years'

all_files = glob.glob(os.path.join(path,"*.csv"))
array = []

for filename in all_files:
    data = pd.read_csv(filename, sep=",", index_col=None, header=0)
    data['Name'] = filename
    array.append(data)
    print(filename)
```

```python
merge_values = np.vstack(array)
data = pd.DataFrame(merge_values)
data.columns = ["Index","Year","Period","%_Vol_Mean","%_GainLose_Mean"]
data = data.rename(columns={'Index':'Year','Year':'Period','Period':'%_Vola_Mean','%_Vol_Mean':'%_GainLose_mean'})
data = data.drop(['%_GainLose_Mean'], axis=1)
```

```python
dataday = data.loc[(data['Period']=='Day')]
dataweek = data.loc[(data['Period']=='Week')]
datamonth = data.loc[(data['Period']=='Month')]
dataday
```

```python
plt.plot(dataday['Year'], dataday['%_Vola_Mean'], label='Volatility')    
plt.plot(dataday['Year'], dataday['%_GainLose_mean'], label='GainLose')

plt.xlabel('Election Years')
plt.ylabel('%')

plt.title("Day to day behavior")
plt.legend()
plt.show()
```

```python
plt.plot(dataweek['Year'], dataweek['%_Vola_Mean'], label='Volatility')    
plt.plot(dataweek['Year'], dataweek['%_GainLose_mean'], label='GainLose')

plt.xlabel('Election Years')
plt.ylabel('%')

plt.title("Week to week behavior")
plt.legend()
plt.show()
```

```python
plt.plot(datamonth['Year'], datamonth['%_Vola_Mean'], label='Volatility')    
plt.plot(datamonth['Year'], datamonth['%_GainLose_mean'], label='GainLose')

plt.xlabel('Election Years')
plt.ylabel('%')

plt.title("Month to month behavior")
plt.legend()
plt.show()
```
