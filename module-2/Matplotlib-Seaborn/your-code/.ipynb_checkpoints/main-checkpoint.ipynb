#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
months = fitbit['Months_encoded']
steps = fitbit['Steps']
week_work = fitbit['Work_or_Weekend']

fig, [ax5, ax6] = plt.subplots(1,2,figsize=(12,4),)
ax5.scatter(months, steps)
ax6.scatter(week_work, steps)

ax5.set_title('Steps per Month')
ax6.set_title('Steps per Work/Weekend')
plt.tight_layout()

ax5.tick_params(length=13, width=1)
ax6.tick_params(length=2, width=1)

fig.savefig('figures.png')
# plt.xticks()
# plt.xticks(np.arange(0, 2, step=1))
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
min_lightly_active = fitbit['Minutes Lightly Active']
min_very_active = fitbit['Minutes Very Active']
min_sedentary = fitbit['Minutes Sedentary']

def plotting(x,y,z):
    fig, [ax11, ax22] = plt.subplots(1,2,figsize=(12,4))
    ax11.scatter(x, y)
    ax22.scatter(z, y)

    ax11.set_title('X vs Y')
    ax22.set_title('Z vs Y')
    plt.tight_layout()
    
plotting(min_lightly_active, steps, min_sedentary)
```
