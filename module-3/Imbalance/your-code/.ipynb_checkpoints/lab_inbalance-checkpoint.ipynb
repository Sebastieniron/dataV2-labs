# Inbalanced Classes
## In this lab, we are going to explore a case of imbalanced classes. 


Like we disussed in class, when we have noisy data, if we are not careful, we can end up fitting our model to the noise in the data and not the 'signal'-- the factors that actually determine the outcome. This is called overfitting, and results in good results in training, and in bad results when the model is applied to real data. Similarly, we could have a model that is too simplistic to accurately model the signal. This produces a model that doesnt work well (ever).

### Clean the dataset. How are you going to integrate the time variable? Do you think the step (integer) coding in which it is given is appropriate?

```python
# Your code here
# Checking for null values
data.isna().sum()
```

```python
# Checking for outliers & deleting outliers.

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

data_new = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
data_new['type'].value_counts()
```

```python
# We are dropping categorical columns nameOrig & nameDest since they have too many possibilities. We could try to delete all
# the numbers and keep the letters in front of them and convert it to a dummy variabel (m=1 no m=0) but since we don't know
# what these letters mean we are opting to drop them.

data_new = data_new.drop(columns=['nameOrig','nameDest'])
```

```python
# For the categorical column 'Type' we will make a distinction between cash and no cash transfers, because my intuition tells me
# cash is the most sketchy form of payment when it comes to trying to detect fraud.

def cash_function(string):
    cashcount = 0
    if "CASH" in (string):
        cashcount =+ 1
    
    return cashcount

data_new['type'] = data_new['type'].apply(lambda row: cash_function(row))
data_new
```

### Run a logisitc regression classifier and evaluate its accuracy.

```python
# Your code here

# https://www.kaggle.com/rafjaa/resampling-strategies-for-imbalanced-datasets

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X = data_new.drop(columns=['isFraud'])
y = data_new['isFraud']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)

model = LogisticRegression()
model.fit(X_train, y_train)

model.score(X_train, y_train)

# ValueError: This solver needs samples of at least 2 classes in the data, but the data contains only one class: 0
# STACKOVERFLOW 
# The problem here is that your y_train vector, for whatever reason, only has zeros. It is actually not your fault, 
# and its kind of a bug ( I think ). The classifier needs 2 classes or else it throws this error.
# It makes sense. If your y_train vector only has zeros, ( ie only 1 class ), then the classifier doesn't really need 
# to do any work, since all predictions should just be the one class.
```
