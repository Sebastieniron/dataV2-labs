```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataLog = pd.read_csv('DataWithLog.csv')
dataLog
```

```python
from sklearn.preprocessing import StandardScaler

standardscaler = StandardScaler()

StandardData = standardscaler.fit(dataLog.drop(columns=['area','day','month']))
StandardData = standardscaler.transform(dataLog.drop(columns=['area','month','day']))
StandardData = pd.DataFrame(StandardData, columns=['FFMC','DMC','DC','ISI','temp','RH','wind','rain'])
StandardData['day'] = dataLog['day']
StandardData['month'] = dataLog['month']
StandardData['area'] = dataLog['area']
StandardData
```

```python
def OrderArea(number):
    if (number>0) and (number<1):
        number = 1
    elif (number>=1) and (number<2):
        number = 2
    elif (number>=2) and (number<3):
        number = 3
    elif (number>=3) and (number<4):
        number = 4
    elif (number>=4) and (number<5):
        number = 5
    elif (number>5):
        number = 6
    
    return number

```

```python
StandardData['area'] = StandardData['area'].apply(lambda row: OrderArea(row))
StandardData
```

```python
StandardData['area'].value_counts()

# The reason why we went until 6 is because there are not enough values over 7. This would cause our SMOTE library to not
# work properly. It needs atleast 5 values per possible outcome.
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.preprocessing import LabelEncoder

X = StandardData.drop(columns=['area'])
y = StandardData['area']

oversample = SMOTE()
X, y = oversample.fit_resample(X, y)

counter = Counter(y)
print(counter)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)

smote_model = KNeighborsClassifier(n_neighbors=3)
smote_model.fit(X_train, y_train)
smote_model.score(X_train, y_train)
```

```python
smote_model.score(X_test, y_test)

# Overfitting?
```

```python
from sklearn.model_selection import cross_val_score

crossvalscore = cross_val_score(KNeighborsClassifier(n_neighbors=3), X, y, cv=20)
print(crossvalscore)
np.mean(crossvalscore)
```
